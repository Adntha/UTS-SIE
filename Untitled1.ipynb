```python
# untuk membaca file data yang sudah diupload di github
source = "UTS-SIE/sales_data_MidTermExam.csv"
data = pd.read_csv(source)
```

```python
data.info()
```

```python
# untuk mencari missing value
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# digunakan untuk menghilangkan missing value pada kolom Age
data.dropna(subset=['Age'], inplace=True)
```

```python
# digunakan untuk mengubah tipe data pada kolom Unit Price menjadi numeric
data['Unit Price'] = data['Unit Price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()
data['Unit Price'] = pd.to_numeric(data['Unit Price'])
```

```python
# Convert 'Date' column to datetime objects
data['Date'] = pd.to_datetime(data['Date'])

# used to create a total column
snapshot_date = max(data['Date']) + pd.DateOffset(days=1)

data['Total'] = data['Units Sold'] * data['Unit Price']
```

```python
# menggunakan customer id sebagai patokan
# menggunakan date sebagai patokan Recency
# menggunakan Transaction ID sebagai patokan Frequency
# menggunakan Total sebagai patokan MoneytaryValue
rfm = data.groupby('Customer ID').agg({
    'Date': lambda x: (snapshot_date - x.max()).days,
    'Transaction ID': 'nunique',
    'Total': 'sum'
})
```

```python
# membuat sebuah kolon rfm
rfm.rename(columns={
    'Date': 'Recency',
    'Transaction ID': 'Frequency',
    'Total': 'MoneytaryValue'
}, inplace=True)

# menampilkan kolom yang sbaru dibuat
rfm.head()
```

```python
# digunakan untuk mengolah data rfm
rfm.describe()
```

```python
# digunakan untuk mengelompokkan score beradasarkan rentangnya
recency_bins = [rfm['Recency'].min()-1,20,50,150,250, rfm['Recency'].max()]
frequency_bins = [rfm['Frequency'].min()-1, 2, 3, 10, 20, rfm['Frequency'].max()]
monetary_bins = [rfm['MoneytaryValue'].min()-3, 300, 600, 2000, 5000, rfm['MoneytaryValue'].max()]

print("uncorrected Recency Bins:", recency_bins)
print("uncorrected Frequency Bins:", frequency_bins)
print("uncorrected MonetaryValue Bins:", monetary_bins)

# Sort the bins in ascending order to ensure they are monotonically increasing
recency_bins.sort()
frequency_bins.sort()
monetary_bins.sort()

print("Corrected Recency Bins:", recency_bins)
print("Corrected Frequency Bins:", frequency_bins)
print("Corrected MonetaryValue Bins:", monetary_bins)
```

```python
#calculate recency score based on custom bins
rfm['R_score'] = pd.cut(rfm['Recency'], bins=recency_bins, labels=range(1,6), include_lowest=True)

#reverse the recency scores so that higher values indicate more recent purchases
rfm['R_score'] = 5 - rfm['R_score'].astype(int) + 1

#calculate  frequency and moneyatry scores based on custom bins
rfm['F_score'] = pd.cut(rfm['Frequency'], bins=frequency_bins, labels=range(1,6), include_lowest=True)
rfm['M_score'] = pd.cut(rfm['MoneytaryValue'], bins=monetary_bins, labels=range(1,6), include_lowest=True)
```
